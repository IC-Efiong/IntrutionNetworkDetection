```python
#Converting float to integers
def flt64_to_int64(df1, col_names):
    for col_name in col_names:
        df1[col_name] = df1[col_name].astype('int64')
    return df1
```

```python
df1 = flt64_to_int64(df1, ['TotLen_Fwd_Pkts', 'TotLen_Bwd_Pkts', 'Bwd_IAT_Tot', 
                            'Fwd_Pkts/s', 'Bwd_Pkts/s'])
```

```python
#Replacing and converting non-finite values to Integers
def float64_to_int64(df1, col_name):
    # handle non-finite values by replacing with 0
    df1[col_name] = df1[col_name].replace([np.inf, -np.inf, np.nan], 0)
    # convert to int64
    df1[col_name] = df1[col_name].astype('int64')
    return df1
```

```python
df1 = float64_to_int64(df1, ['Flow_Byts/s', 'Flow_Pkts/s'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#function to encode categorical variables
def le_encoder(data):
    for col in df1.columns:
        if df1[col].dtype == 'object':
                label_encoder = LabelEncoder()
                df1[col] = label_encoder.fit_transform(df1[col])
```

# Feature Selection

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
```

```python
X = df1.drop(labels='Cat', axis=1)
y = df1['Cat']
```

```python
#feature selection using Mutual Information technique 
mutual_info = mutual_info_classif(X,y)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X.columns
mutual_info.sort_values(ascending=False)
```

```python
#plottiing features score
mutual_info.sort_values(ascending=False).plot.bar(figsize=(20,8))
```

```python
#selecting the best columns for building the model
select_best_cols = SelectKBest(mutual_info_classif,k=29)
select_best_cols.fit(X,y)
selected_features = X.columns[select_best_cols.get_support()]
```

```python
X=X[selected_features]
```

# Split, Scale and Fit Data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#Spliting the datset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
scaler = StandardScaler()
```

```python
#Scaling the train and test data
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Model Building

```python
#model building
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,BaggingClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report,ConfusionMatrixDisplay,\
                            precision_score, recall_score, f1_score, roc_auc_score,roc_curve,confusion_matrix
```

### Logistic Regression Algorithm

```python
lr = LogisticRegression() #creating the model
lr.fit(X_train,y_train) #fiting the dataset into the model
lr_y_train_pred = lr.predict(X_train) #predicting train dataset
lr_y_test_pred = lr.predict(X_test) #predicting test dataset
lr_train_accuracy = accuracy_score(y_train, lr_y_train_pred) #accuracy score for train dataset
lr_test_accuracy = accuracy_score(y_test, lr_y_test_pred) #accuracy score for test dataset
lr_classification = classification_report(y_test, lr_y_test_pred) #classification report

print()
print('============================== Logistic Regression Model Evaluation ==============================')
print()
print("Train Model Accuracy:" "\n",lr_train_accuracy)
print()
print("Test Model Accuracy:" "\n",lr_test_accuracy)
print()
print("Classification report:" "\n", lr_classification)
print()        
```

### Random Forest Algorithm

```python
rfc = RandomForestClassifier(n_estimators= 10, criterion="entropy") #creating the model
rfc.fit(X_train,y_train) #fiting the dataset into the model
rfc_y_train_pred = rfc.predict(X_train) #predicting train dataset
rfc_y_test_pred = rfc.predict(X_test) #predicting test dataset
rfc_train_accuracy = accuracy_score(y_train, rfc_y_train_pred) #accuracy score for train dataset
rfc_test_accuracy = accuracy_score(y_test, rfc_y_test_pred) #accuracy score for test dataset
rfc_classification = classification_report(y_test, rfc_y_test_pred) #classification report

print()
print('============================== Random Forest Classifier Model Evaluation ==============================')
print()
print("Train Model Accuracy:" "\n",rfc_train_accuracy)
print()
print("Test Model Accuracy:" "\n",rfc_test_accuracy)
print()
print("Classification report:" "\n", rfc_classification)
print()          
```
